```python
from sklearn.ensemble import VotingClassifier
from sklearn.datasets import make_classification,make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
X,y=make_classification(
    n_samples=500,
    n_features=20,
    n_informative=5,
    n_redundant=2,
    random_state=42
)

x_train,x_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
lr=LogisticRegression()
svc=SVC()
dtc=DecisionTreeClassifier(max_depth=3)
```

```python
voting_clf=VotingClassifier(
    estimators=[('lr',lr),('svc',svc),('dtc',dtc)],
    voting='hard',
    # n_weights=
)
```

```python
voting_clf.fit(x_train,y_train)
lr.fit(x_train,y_train)
svc.fit(x_train,y_train)
dtc.fit(x_train,y_train)
```

```python
y_pred=voting_clf.predict(x_test)
y_pred1=lr.predict(x_test)
y_pred2=svc.predict(x_test)
y_pred3=dtc.predict(x_test)
```

```python
print('accuracy',accuracy_score(y_test,y_pred))
print('cr',classification_report(y_test,y_pred))
print('accuracy',accuracy_score(y_test,y_pred1))
print('cr',classification_report(y_test,y_pred))
print('accuracy',accuracy_score(y_test,y_pred))
print('cr',classification_report(y_test,y_pred))
print('accuracy',accuracy_score(y_test,y_pred))
print('cr',classification_report(y_test,y_pred))
```

```python
X,y=make_regression(
    n_samples=500,
    n_features=20,
    n_informative=5,
    random_state=42
)

x_train,x_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
```

```python
lr=LinearRegression()
svr=SVR()
dtc=DecisionTreeRegressor(max_depth=3)
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr=VotingRegressor(
    estimators=[
        ('lr',lr),
        ('svr',svr),
        ('dtc',dtc)
    ]
)
```

```python
vr.fit(x_train,y_train)
lr.fit(x_train,y_train)
svr.fit(x_train,y_train)
dtc.fit(x_train,y_train)
```

```python
y_pred=vr.predict(x_test)
y_pred1=lr.predict(x_test)
y_pred2=svr.predict(x_test)
y_pred3=dtc.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('vr',r2_score(y_test,y_pred))
print('lr',r2_score(y_test,y_pred1))
print('svr',r2_score(y_test,y_pred2))
print('dtc',r2_score(y_test,y_pred3))
```
