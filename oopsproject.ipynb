```python
class NLPApp:
    def __init__(self):
        
        
        self.__database={}
        self.__first_menu()
    def get(self):
        return self.__database
        
    def __first_menu(self):
        
        first_input=input("""
        Hi! how would you like to proceed?
        1.not member? register
        2. login
        3.exit""")
        if first_input=='1':
            self.__register()
        elif first_input=='2':
            self.__login()
        else:
            exit
            
    def __register(self):
        name=input('enter the name:')
        email=input('enter the email:')
        password=input('enter the password:')
        if email in self.__database:
            print('email already exist')
        else:
            self.__database[email]=[name,password] 
            print('registration successful')
            self.__first_menu()
            
    def __login(self):
        email=input('enter the email:')
        name=input('enter the name:')
        password=input('enter the password:')
        if email in self.__database:
            if self.__database[email]==[name,password]:
                self.__second_menu()
            else:
                print('wrong password')
        else:
            print('wrong email')
                    
                
                
                
              
            
    def __second_menu(self):
        second_input=input("""
        1.NER
        2.sentiment analysis
        3.logout""")
        if second_input=='1':
            self.__NER()
      
        elif second_input=='2':
            self.__sentiment_analysis()
        else:
            self.__logout()
    
    
    def __NER(self):
        para=input('enter the para:')
        search=input('search element:')
        

        client = nlpcloud.Client("gpt-oss-120b", "36ef6a69a54fb4276047511dbc3e4d6ba4d43c0e", gpu=True)
        response=client.entities(para,
                        searched_entity=search)
        print(response)
        
    def __sentiment_analysis(self):
        lang_text=input('enter the text in different languages')
        

        

        client = nlpcloud.Client("distilbert-base-uncased-emotion", "2b58d7fb9af09e617ee525e78c7766b6d8c5bb61", gpu=False, lang="en")
        response=client.sentiment(lang_text)
        print(response)
        
    def __logout(self):
        exit

        

obj=NLPApp()

   
```
